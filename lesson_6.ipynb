```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

```python
def read_text(filename):
    f = open(filename, "r")
    with f:
        results = []
        lines = f.readlines()
        for line in lines[1:]:
            x, y = line.split(" ")
            results.append([int(x), float(y)])
        return results
    
    
def annotate_max(x, y):
    ymax = max(y)
    xpos = y.index(ymax)
    xmax = x[xpos]
    plt.annotate('max value', xy=(xmax, ymax), xytext=(xmax+10, ymax),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )

    
```

```python
data = read_text("dataset.txt")

#plot 1 - liniar
fig = plt.figure()
fig.set_figwidth(30)
fig.set_figheight(15)
plt.subplot(2, 2, 1)
plt.plot([x[0] for x in data], [x[1] for x in data])
plt.plot([x[0] for x in data], [2 * x[1] for x in data])
annotate_max(x, y)

#plot 2 - step
fig = plt.figure()
fig.set_figwidth(30)
fig.set_figheight(15)
plt.subplot(2, 2, 2)
plt.step([x[0] for x in data], [x[1] for x in data])
plt.step([x[0] for x in data], [2 * x[1] for x in data])
annotate_max(x, y)

#plot 3 - stem
fig = plt.figure()
fig.set_figwidth(30)
fig.set_figheight(15)
plt.subplot(2, 2, 3)
plt.stem([x[0] for x in data], [x[1] for x in data], "--")
plt.stem([x[0] for x in data], [2 * x[1] for x in data], ":")
annotate_max(x, y)

#plot 4 - bar
fig = plt.figure()
fig.set_figwidth(30)
fig.set_figheight(15)
width = 0.40
plt.subplot(2, 2, 4)
plt.bar([x[0]-0.3 for x in data], [x[1] for x in data], width)
plt.bar([x[0]+0.3 for x in data], [2 * x[1] for x in data], width)
annotate_max(x, y)

plt.show
```
